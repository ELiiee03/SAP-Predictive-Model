# Univariate analysis

```python
# Count plots for categorical features
categorical_features = ['philhealth_status', 'ip', 'gender', 'civil_status', 'highest_education_attainment', 'osy_isy', 'out', 'youth_with_specific_needs', 'disability', 'work_status']

for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=dataset, y=feature)
    plt.title(f'Count Plot for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
# Plot histograms for numerical features
numerical_features = ['age', 'number_of_children']

for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    plt.hist(dataset[feature], bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Histogram for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

# Bivariate analysis

The count plots generated for the categorical and numerical features in the dataset provide a comprehensive visual analysis of how these features are being distributed across different priority levels for social assistance. Each plot represents the distribution of a specific categorical feature, such as philhealth_status, ip, gender, age, civil_status, number_of_children, highest_education_attainment, osy_isy, out, youth_with_specific_needs, disability, and work_status, with the bars colored according to the priority_level.

```python
# Count plots for categorical features
categorical_features = ['philhealth_status', 'ip', 'civil_status', 'gender', 'osy_isy', 'youth_with_specific_needs', 'disability', 'work_status']

for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=dataset, x=feature, hue='priority_level')
    plt.title(f'Distribution of {feature} by Priority Level')
    plt.show()
```

```python
# Plot for 'out' by 'priority_level'
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, y='out', hue='priority_level')
plt.title('Distribution of out by Priority Level')
plt.xlabel('Philhealth Status')
plt.ylabel('Count')
plt.show()

# Plot for 'highest_education_attainment' by 'priority_level'
plt.figure(figsize=(8, 10))
sns.countplot(data=dataset, y='highest_education_attainment', hue='priority_level')
plt.title('Distribution of highest_education_attainments by Priority Level')
plt.xlabel('Philhealth Status')
plt.ylabel('Count')
plt.show()
```

```python
# Numerical category
# Box plot for 'age' by 'priority_level'
plt.figure(figsize=(8, 5))
sns.boxplot(data=dataset, x='priority_level', y='age')
plt.title('Distribution of Age by Priority Level')
plt.xlabel('Priority Level')
plt.ylabel('Age')
plt.show()

# Numerical category
# Box plot for 'number_of_children' by 'priority_level'
plt.figure(figsize=(8, 6))
sns.boxplot(data=dataset, x='priority_level', y='number_of_children')
plt.title('Distribution of Number of Children by Priority Level')
plt.xlabel('Priority Level')
plt.ylabel('Number of Children')
plt.show()
```

# Multivariate analysis

Strong Correlations (key features):
1. priority_level and ip (0.52): There is a moderate positive correlation between being an ip (likely referring to Indigenous Peoples or a specific group) and being a priority. This suggests that individuals from this group are more likely to be prioritized for financial assistance.

2. priority_level and osy_isy (0.47): A moderately strong positive correlation indicates that being an out-of-school youth or in-school youth (osy_isy) is a significant predictor of priority level. This group likely requires more assistance.

3. priority_level and age (0.22):A weaker positive correlation suggests that age might have some influence on prioritization. Younger or older individuals might have a slightly higher likelihood of being prioritized.

4. priority_level and number_of_children (0.19): A positive correlation indicates that households with more children are more likely to be classified as a priority, which aligns with the assumption that larger families may have greater financial needs.

Weak or No Correlations:
1. priority_level and philhealth_status (0.037): PhilHealth status shows almost no correlation with priority level, despite its potential importance. This might indicate that the dataset's prioritization criteria do not strongly consider health insurance coverage.

2. priority_level and disability (0.028): The very weak correlation suggests that disability status is not a significant factor in determining priority. This might warrant further investigation to assess whether this aligns with policy goals.

3. priority_level and highest_education_attainment (0.076): Education level has minimal correlation with priority, suggesting it is not a major factor for determining financial assistance in this dataset.

4. priority_level and gender (-0.07): The negative correlation indicates that gender has almost no influence on the priority level.
